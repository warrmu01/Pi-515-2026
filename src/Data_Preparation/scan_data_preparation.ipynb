```python
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.model_selection import train_test_split
```

```python
import sys
import os

sys.path.append(os.path.abspath(".."))
sys.path.append(os.path.abspath("../.."))
```

```python
BASE_DIR = Path.cwd().parents[1]
RAW_DATA_PATH = BASE_DIR / "Data" / "SCAN" / "raw" / "SCAN_ames_all.csv"
PROCESSED_DIR = BASE_DIR / "Data" / "SCAN" / "processed"
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

print("BASE_DIR:", BASE_DIR)
print("RAW_DATA_PATH exists:", RAW_DATA_PATH.exists())
print("PROCESSED_DIR:", PROCESSED_DIR)
```

```python
scan_df = pd.read_csv(RAW_DATA_PATH, comment="#")
scan_df.head()
```

```python
print(scan_df.columns.tolist())
print(scan_df.shape)
```

```python
selected_columns = {
    "Date": "date",
    "Ames (2031) Precipitation Increment (in)": "total_precipitation",
    "Ames (2031) Air Temperature Average (degF)": "air_temperature",
    "Ames (2031) Dew Point Temperature (degF)": "dewpoint_temperature",
    "Ames (2031) Soil Moisture Percent -2in (pct)": "soil_moisture",
    "Ames (2031) Soil Temperature Observed -2in (degF)": "soil_temperature",
}
```

```python
scan_df = scan_df[list(selected_columns.keys())].rename(columns=selected_columns)
scan_df.head()
```

```python
scan_df = scan_df.replace(r"^\s*$", np.nan, regex=True)
scan_df = scan_df.dropna()

print(scan_df.shape)
scan_df.head()
```

```python
scan_df["date"] = pd.to_datetime(scan_df["date"])

numeric_cols = [
    "total_precipitation",
    "air_temperature",
    "dewpoint_temperature",
    "soil_moisture",
    "soil_temperature",
]

for col in numeric_cols:
    scan_df[col] = pd.to_numeric(scan_df[col], errors="coerce")

scan_df = scan_df.dropna()

scan_df.dtypes
```

```python
# Fahrenheit -> Celsius
scan_df["air_temperature"] = (scan_df["air_temperature"] - 32) * 5 / 9
scan_df["dewpoint_temperature"] = (scan_df["dewpoint_temperature"] - 32) * 5 / 9
scan_df["soil_temperature"] = (scan_df["soil_temperature"] - 32) * 5 / 9

# percent -> proportion
scan_df["soil_moisture"] = scan_df["soil_moisture"] / 100

scan_df.head()
```

```python
scan_df["year"] = scan_df["date"].dt.year
scan_df["month"] = scan_df["date"].dt.month
scan_df["day"] = scan_df["date"].dt.day

scan_df.head()
```

```python
scan_df["latitude"] = 42.0347
scan_df["longitude"] = -93.62

scan_df.head()
```

```python
daily_scan_df = (
    scan_df.set_index("date")
    .resample("D")
    .agg({
        "total_precipitation": "sum",
        "air_temperature": "mean",
        "dewpoint_temperature": "mean",
        "soil_moisture": "mean",
        "soil_temperature": "mean",
        "year": "first",
        "month": "first",
        "day": "first",
        "latitude": "first",
        "longitude": "first",
    })
    .reset_index()
)

daily_scan_df.head()
```

```python
final_scan_df = daily_scan_df.copy()
```

```python
final_scan_df["source"] = "SCAN"
final_scan_df.head()
```

```python
from sklearn.preprocessing import StandardScaler
import joblib

feature_cols_to_scale = [
    "latitude",
    "longitude",
    "air_temperature",
    "dewpoint_temperature",
    "total_precipitation",
    "year",
    "month",
    "day",
]

scan_scaler = StandardScaler()
final_scan_df[feature_cols_to_scale] = scan_scaler.fit_transform(
    final_scan_df[feature_cols_to_scale]
)

MODEL_DIR = BASE_DIR / "src" / "models"
MODEL_DIR.mkdir(parents=True, exist_ok=True)

joblib.dump(scan_scaler, MODEL_DIR / "scan_feature_scaler.joblib")

print("✅ Standardized SCAN numeric features")
print("✅ Saved SCAN scaler to:", MODEL_DIR / "scan_feature_scaler.joblib")

final_scan_df.head()
```

```python
final_scan_df[feature_cols_to_scale].describe()
```

```python
required_cols = [
    "latitude",
    "longitude",
    "air_temperature",
    "dewpoint_temperature",
    "total_precipitation",
    "year",
    "month",
    "day",
    "soil_temperature",
    "soil_moisture",
]

final_scan_df = final_scan_df.dropna(subset=required_cols).copy()

print("Final SCAN shape after dropping nulls:", final_scan_df.shape)
print(final_scan_df[required_cols].isnull().sum())
```

```python
scan_train_df, scan_holdout_df = train_test_split(
    final_scan_df,
    test_size=0.10,
    random_state=42,
    shuffle=True
)

print("SCAN train shape:", scan_train_df.shape)
print("SCAN holdout shape:", scan_holdout_df.shape)
```

```python
final_scan_df.to_csv(PROCESSED_DIR / "scan_processed.csv", index=False)
scan_train_df.to_csv(PROCESSED_DIR / "scan_train.csv", index=False)
scan_holdout_df.to_csv(PROCESSED_DIR / "scan_holdout.csv", index=False)

print("✅ Saved:")
print(PROCESSED_DIR / "scan_processed.csv")
print(PROCESSED_DIR / "scan_train.csv")
print(PROCESSED_DIR / "scan_holdout.csv")
```
