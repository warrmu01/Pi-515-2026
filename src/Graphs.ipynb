```python
import matplotlib.pyplot as plt

# Clean style
plt.rcParams.update({
    "figure.figsize": (9, 5),
    "axes.titlesize": 18,
    "axes.labelsize": 13,
    "xtick.labelsize": 11,
    "ytick.labelsize": 12,
    "axes.spines.top": False,
    "axes.spines.right": False,
})

# Dark magenta color
MAGENTA = "#7A1C4A"
```

```python
labels = ["SMOS-only", "Combined (SMOS)", "Combined (SCAN)"]
values = [0.9839, 0.9768, 0.9633]

plt.figure()
bars = plt.barh(labels, values, color=MAGENTA, height=0.5)

plt.title("Soil Temperature Model Comparison (R²)", pad=15)
plt.xlabel("R²")

# Add value labels
for bar, value in zip(bars, values):
    plt.text(
        bar.get_width() + 0.002,
        bar.get_y() + bar.get_height()/2,
        f"{value:.4f}",
        va='center',
        fontsize=11
    )

plt.xlim(0.95, 1.0)
plt.grid(axis='x', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.savefig("soil_temp_comparison.png", dpi=300, bbox_inches="tight")
plt.show()
```
